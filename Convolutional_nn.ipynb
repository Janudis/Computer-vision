```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader

from torchvision import datasets
from torchvision.transforms import ToTensor

from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

```python
# device for all computations
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)
```

```python
# Το δίκτυο που πρέπει να σχεδιάσετε
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        # εδώ γράφουμε τα επίπεδα του δικτύου μας
        self.conv0 = nn.Conv2d(1, 8, 7) #1 gia grayscale, 8 arithmos filtron, 7 to megethos
        self.conv1 = nn.Conv2d(8, 32, 5)
        self.fc0 = nn.Linear(288, 128) #288=Height*Weight*Depth apo x = torch.flatten(x, start_dim=1) 
        self.fc1 = nn.Linear(128, 50)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = F.relu(self.conv0(x))
        x = F.max_pool2d(x, kernel_size=2, stride=2) #megethos filtrou 2 me vima 2
        
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, kernel_size=2, stride=2)
        
        x = torch.flatten(x, start_dim=1)
        print(x.shape)
        
        x = F.relu(self.fc0(x))
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = F.log_softmax(x, dim=1)
        return x


class Trainer():
    def __init__(self, model, train_dataset, test_dataset, batch_size):
        # device to put tensors on
        self.device = torch.device(
            "cuda" if torch.cuda.is_available() else "cpu"
        )
        # model
        self.model = model.to(self.device)
        # dataloaders
        self.train_loader = DataLoader(train_dataset,
                                       batch_size=batch_size,
                                       shuffle=True)
        self.test_loader = DataLoader(test_dataset,
                                      batch_size=batch_size,
                                      shuffle=True)
        # optimizer
        self.optimizer = optim.Adam(self.model.parameters(), lr=0.001)

    def count_parameters(self):
        params = sum((p.numel() for p in self.model.parameters()
                     if p.requires_grad))
        return params

    def train(self):
        # set model to train mode
        self.model.train()

        # variable to store all batch losses
        epoch_losses = 0

        # train one epoch
        for batch, label in self.train_loader:
            # move tensors to device
            batch = batch.to(self.device)
            label = label.to(self.device)

            # clear gradients
            self.optimizer.zero_grad()
            # forward pass
            output = self.model(batch)
            # calculate loss
            loss = F.nll_loss(output, label)
            # backpropagation
            loss.backward()
            # optimization step
            self.optimizer.step()

            epoch_losses += loss.item()

        # mean epoch loss
        epoch_losses /= len(self.train_loader)

        return epoch_losses

    @torch.no_grad()
    def evaluate(self):
        # evaluation mode
        self.model.eval()

        test_losses = 0
        test_accuracy = 0

        for batch, label in self.test_loader:
            # move tensors to device
            batch = batch.to(self.device)
            label = label.to(self.device)

            # forward pass
            output = self.model(batch)
            # calculate test loss
            loss = F.nll_loss(output, label)
            test_losses += loss.item()

            # calculate test accuracy
            preds = output.argmax(1).cpu().numpy()
            label = label.cpu().numpy()
            accuracy = accuracy_score(label, preds)
            test_accuracy += accuracy

        # mean test loss / mean test accuracy
        test_losses /= len(self.test_loader)
        test_accuracy /= len(self.test_loader)

        return test_losses, test_accuracy

    def run(self, epochs):
        print(self.model, "\n")

        params = self.count_parameters()
        print(f"Total Parameters: {params}")
        print(f"Using Device: {self.device}")

        train_losses = []
        test_losses = []

        test_accuracy = []

        print("\nTraining...")

        for epoch in range(epochs):
            print(f"\nEpoch: {epoch:2d}")

            train_loss = self.train()
            print(f"Train Loss: {train_loss:.4f}")

            test_loss, test_acc = self.evaluate()
            print(f"Test Loss:  {test_loss:.4f} \tTest Accuracy: {test_acc:.4f}")

            train_losses.append(train_loss)
            test_losses.append(test_loss)
            test_accuracy.append(test_acc)

        return train_losses, test_losses, test_accuracy


def plot_stats(train_loss, test_loss, accuracy):
    fig, (ax1, ax2) = plt.subplots(2)

    ax1.plot(train_loss, label="Train Loss")
    ax1.plot(test_loss, label="Evaluation Loss")
    ax1.set_title("Losses")
    ax1.set(xlabel="Epoch", ylabel="")
    ax1.legend()
    ax1.label_outer()

    ax2.plot(accuracy)
    ax2.set_title("Evaluation Accuracy")
    ax2.set(xlabel="Epoch", ylabel="")
    ax2.label_outer()

    plt.show()


if __name__ == "__main__":
    # dataset
    # for training
    train_dataset = datasets.FashionMNIST(root="data",
                                          train=True,
                                          download=True,
                                          transform=ToTensor())
    # for test / evaluation
    test_dataset = datasets.FashionMNIST(root="data",
                                         train=False,
                                         download=True,
                                         transform=ToTensor())

    # hyperparameters
    epochs = 15
    batch_size = 128

    # initialize model
    model = CNN()

    # create trainer object
    trainer = Trainer(model, train_dataset, test_dataset, batch_size)

    # train model and get losses and test accuracy
    train_loss, test_loss, accuracy = trainer.run(epochs)

    # plot losses and accuracy
    plot_stats(train_loss, test_loss, accuracy)
```
